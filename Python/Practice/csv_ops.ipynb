```python
import csv

# Reading a CSV file into a list of dictionaries
with open('/Workspace/Users/pundpaltejas@gmail.com/DE-Practice/Python/Practice/customer.csv', mode='r', newline='') as file:
    reader = csv.DictReader(file)
    data = list(reader)

# Display all rows
for row in data:
    print(row)

# Writing data to a CSV file
with open('/Workspace/Users/pundpaltejas@gmail.com/DE-Practice/Python/Practice/customer_output.csv', mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=['id', 'name', 'address'])
    writer.writeheader()
    writer.writerows(data)

# Filtering rows where address is 'India'
filtered_data = [row for row in data if row['address'] == 'India']
for row in filtered_data:
    print(row)

# Selecting specific columns
selected_data = [{'id': row['id'], 'name': row['name']} for row in data]
for row in selected_data:
    print(row)

# Aggregating data (count by address)
from collections import Counter
address_counts = Counter(row['address'] for row in data)
for address, count in address_counts.items():
    print({'address': address, 'count': count})
```
