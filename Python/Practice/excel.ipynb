```python
import openpyxl

wb = openpyxl.load_workbook('/Workspace/Users/pundpaltejas@gmail.com/DE-Practice/Python/Practice/example_excel_file.xlsx')
type(wb)
```

```python
wb.sheetnames
```

```python
sheet1 = wb['Sales Data']
sheet1
print(sheet1.title)

active_sheet = wb.active
print(active_sheet.title)

print(sheet1['A1'].value)

c = sheet1['B1']
print(c.value)
print(c.row)
print(c.column)
f'Row {c.row}, Column {c.column} is {c.value}'
```

```python
#max rows and columns
print(sheet1.max_row)
print(sheet1.max_column)
```

```python
#iterating throght all rows and columns
for row in sheet1.iter_rows(min_row=1, max_row=11, min_col=1, max_col=4):
  for cell in row:
    print(cell.value, end=' ')
  print()
```

```python
print('Opening workbook...')
sheet = wb['Employee Info']
employee_data = {}

print('Reading rows...')
for row in range(2, sheet.max_row + 1):
    employee_id = sheet.cell(row=row, column=1).value
    name = sheet.cell(row=row, column=2).value
    department = sheet.cell(row=row, column=3).value
    salary = sheet.cell(row=row, column=4).value
    employee_data[employee_id] = (name, department, salary)
    
    print(employee_id, employee_data[employee_id])
```
