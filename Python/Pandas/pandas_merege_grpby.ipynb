```python
import pandas as pd

# Create employee DataFrame
employee_data = [
    {'id': 1, 'name': 'Alice', 'salary': 90000, 'departmentId': 101},
    {'id': 2, 'name': 'Bob', 'salary': 85000, 'departmentId': 101},
    {'id': 3, 'name': 'Charlie', 'salary': 95000, 'departmentId': 102},
    {'id': 4, 'name': 'David', 'salary': 95000, 'departmentId': 102},
    {'id': 5, 'name': 'Eve', 'salary': 80000, 'departmentId': 103}
]
employee = pd.DataFrame(employee_data)

# Create department DataFrame
department_data = [
    {'id': 101, 'name': 'HR'},
    {'id': 102, 'name': 'Engineering'},
    {'id': 103, 'name': 'Marketing'}
]
department = pd.DataFrame(department_data)

# Merge employee and department DataFrames on departmentId
merged_df = pd.merge(employee, department, left_on='departmentId', right_on='id', how='inner')

display(employee)
display(department)
display(merged_df)

# Highest salary per department
highest_salary = merged_df.groupby('departmentId')['salary'].max().reset_index(name='highest_salary')
display(highest_salary)

# Second highest salary per department
def second_highest(s):
    unique_salaries = s.unique()
    if len(unique_salaries) > 1:
        return pd.Series(sorted(unique_salaries, reverse=True)[1])
    else:
        return pd.Series([None])

second_highest_salary = merged_df.groupby('departmentId')['salary'].apply(second_highest).reset_index(name='second_highest_salary')
display(second_highest_salary)

# Average salary per department
avg_salary = merged_df.groupby('departmentId')['salary'].mean().reset_index(name='avg_salary')
display(avg_salary)

# Employees with highest salary per department
def employees_with_highest_salary(df):
    max_salary = df['salary'].max()
    return df.loc[df['salary'] == max_salary, 'id'].tolist()

employees_highest = merged_df.groupby('departmentId').apply(employees_with_highest_salary).reset_index(name='employees_with_highest_salary')
display(employees_highest)

# Employees with second highest salary per department
def employees_with_second_highest_salary(df):
    unique_salaries = sorted(df['salary'].unique(), reverse=True)
    if len(unique_salaries) > 1:
        second_highest = unique_salaries[1]
        return df[df['salary'] == second_highest]['id'].tolist()
    else:
        return []

employees_second_highest = merged_df.groupby('departmentId').apply(employees_with_second_highest_salary).reset_index(name='employees_with_second_highest_salary')
display(employees_second_highest)
```
