Bubble Sort :

Bubble sort is a sorting algorithm that sorts a list of elements iteratively by checking adjacent elements and swapping them if the elements are not in the correct order.

- Best Case: O(n) → When the array is already sorted (with an optimized version that checks for swaps).

- Worst Case: O(n square) → When the array is sorted in reverse order.

```python
def bubbleSort(data):
    size = len(data)
    for i in range(size):
        for j in range(0, size - i - 1):
            if data[j] > data[j + 1]:
                data[j], data[j + 1] = data[j + 1], data[j]
    return data

nums = [5, 3, 8, 6, 7, 2]
print(bubbleSort(nums))
```

Insertion Sort:

The insertion sort algorithm concept is based on the deck of the card where we sort the playing card according to a particular card.

- It doesn't require to know the array size in advance and it receives the one element at a time.

- Best Case: O(n) → When the array is already sorted.

- Worst Case: O(n square) → When the array is sorted in reverse order.

```python
def insertionSort(data):
    size = len(data)
    for i in range(1, size):
        key = data[i]
        j = i - 1
        while j >= 0 and key < data[j]:
            data[j + 1] = data[j]
            j -= 1
        data[j + 1] = key

    return data

nums = [5, 3, 8, 6, 7, 2]
print(insertionSort(nums))
```

Selection Sort :

In this algorithm, we select the smallest element from an unsorted array in each pass and swap with the beginning of the unsorted array. 

- Best Case: O(n square) → No matter the input, it always performs the same number of comparisons.

- Worst Case: O(n square) → Same as best case; not adaptive.

```python
def selectionSort(data):
    size = len(data)
    for i in range(size-1):
        mini = i
        for j in range(i + 1, size):
            if(data[j] < data[mini]):
                mini = j
        data[i], data[mini] = data[mini], data[i]
    return data

num = [5, 3, 1, 10, 9, 2]
print(selectionSort(num))
```

Merge Sort:

It divides the given list in the two halves, calls itself for the two halves and then merges the two sorted halves. We define the merge() function used to merging two halves.

- It is the best example for divide and conquer category of algorithms.

- Best Case: O(n logn) -> Even if the array is already sorted, merge sort still divides and merges recursively

- Wortst Case: O(n logn) -> Merge sort always divides the array into halves and merges them, regardless of input order.

```python
def mergeSort(data):
    if len(data) <= 1:
        return data
    mid = len(data)//2
    left = mergeSort(data[:mid])
    right = mergeSort(data[mid:])
    return merge(left, right)

def merge(left, right):
    result = []
    i,j = 0,0

    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])  
            i += 1
        else:
            result.append(right[j])
            j += 1
    result.extend(left[i:])
    result.extend(right[j:])
    return result

data = [8, 3, 6, 1, 4, 2, 9, 5, 7]
print(mergeSort(data))
```

Quick Sort:

It is based on the divide-and-conquer strategy. Quicksort algorithm works by partitioning an array into two sub-arrays, one containing elements less than a picked pivot element, and the other containing elements more noteworthy than or equivalent to the pivot element.

- Best Case: O(nlogn) → When the pivot divides the array into two equal halves.

- Worst Case: O(n square) → When the pivot is always the smallest or largest element (e.g., sorted or reverse sorted array without randomization).

```python
def quickSort(data,low,high):
    if low < high:
        partition_pos = partition(data,low,high)
        quickSort(data,low,partition_pos-1)
        quickSort(data,partition_pos+1,high)

def partition(data,low,high):
    pivot = data[low]
    i = low + 1
    j = high

    while True:
        while i <=j and data[i] <= pivot:
            i += 1
        while i <= j and data[j] > pivot:
            j -= 1
        if i <= j:
            data[i],data[j] = data[j],data[i]
        else:
            break
    data[low],data[j] = data[j], data[low]

    return j

data = [8, 3, 6, 1, 4, 2, 9, 5, 7, 0]
quickSort(data,0,len(data)-1)
print(data)
```
