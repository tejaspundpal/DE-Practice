###Linear Search:

```python
def linearSearch(data, target):
    size = len(data)
    if size == 0:
        print("Empty list")
        return -1
    for i in range(size):
        if data[i] == target
            return i
    return -1  

data = [4,8,2,1,5,7,6,3]
target = 5
print(f'Target {target} found at index {linearSearch(data, target)}')    
```

###Binary Search

```python
def binarySearch(data, target):
    size = len(data)
    if size == 0:
        print("Empty list")
    left = 0
    right = size - 1
    while left <= right:
        mid = (left + right) // 2
        if target == data[mid]:
            return mid
        elif target < data[mid]:
            right = mid - 1
        else:
            left = mid + 1
    return -1
data = [1,2,3,4,5,6,7,8]
target = 5
print(f'Target {target} found at index {binarySearch(data, target)}')    
```
